```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
```

```python
dummies =pd.get_dummies(df.town)
```

```python
merged = pd.concat([df, dummies], axis ='columns')
merged
```

```python
merged.drop(['town', 'west windsor'], axis = 'columns', inplace = True)
merged
```

```python
from sklearn.linear_model import  LinearRegression
model = LinearRegression()
```

```python
X = merged.drop(['price'], axis = 'columns')
y = merged.price
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```

```python
data = pd.read_csv('adult.data', header =  None, index_col =False, names = ['age','workclass','fnlwgt','education','educ-num','marital','occupation','relationship','race','gender','capital-gain','capital-loss','hours/week','country','income'])
```

```python
data.head()
```

```python
data = data[['age', 'race','education', 'marital', 'country', 'income']]
data.head()
```

```python
print('Features before encoding:\n', list(data.columns),'\n')
print('Original number of features:\n', data.shape[1], '\n')
data_dummies = pd.get_dummies(data)
print('Features after encoding:\n', list(data_dummies.columns))
print('New number of features after encoding:\n', data_dummies.shape[1], '\n')
print('Increase in the number of features:', data_dummies.shape[1]-data.shape[1])
```

```python
features = data_dummies.loc[:, 'age':'country_ Yugoslavia'].values

target = data_dummies.loc[:,'income_ >50K'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, random_state =0)
logreg = LogisticRegression()
logreg.fit(X_train,y_train)

print('Logistic regression score on the test set: {}'. format(logreg.score(X_test, y_test)))
```
